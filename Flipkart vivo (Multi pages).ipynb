```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
Product_Name = []
Product_Price = []

for i in range(1, 13):

    url = f'https://www.flipkart.com/mobiles/pr?sid=tyy%2C4io&otracker=categorytree&p%5B%5D=facets.brand%255B%255D%3DVivo&otracker=nmenu_sub_Electronics_0_Vivo&page={i}'
    response = requests.get(url)

    soup = BeautifulSoup(response.content, 'html.parser')

    names = soup.find_all("div", class_="_4rR01T")

    for a in names:
        var = a.text
        Product_Name.append(var)

    price = soup.find_all("div", class_="_30jeq3 _1_WHN1")

    for a in price:
        var = a.text
        Product_Price.append(var)

df = pd.DataFrame({"Product Name": Product_Name, "Prices": Product_Price})
print(df)
```
